```python
import spacy
import re
import math
import random

nlp_grk = spacy.load("el_core_news_sm")

with open ("greek_text.txt", "r", encoding="utf-8") as f:
    text = f.read()
print(text)
```

# Cleaning the Corpus

```python
#Cleaning the text

clean_text = re.sub("\(.*\)", "", text)
clean_text2 = re.sub("-.*[\.\?]", "", clean_text)

lines = clean_text2.split("\n")

cleaned_lines = []
for line in lines:
    cleaned_line = line.rstrip()
    cleaned_lines.append(cleaned_line)

final_text = " ".join(cleaned_lines)
print(final_text)
```

# Tokenizing the Corpus

```python
#Tokenizing the corpus
doc = nlp_grk(final_text)

for token in doc:
    print(token)

total_words_in_corpus = len(doc)
print(total_words_in_corpus)
```

# Counting Unigrams

```python
#counting unigrams
unigram_count = {}
for token in doc:
    if token.text in unigram_count:
        unigram_count[token.text] += 1
    else:
        unigram_count[token.text] = 1
print(unigram_count)

vocab_size = len(unigram_count)
print(vocab_size) #Vocabulary size
```

# Counting Bigrams

```python
#counting Bigrams
bigram_count = {}
for i in range(len(doc) - 1):
    bigram = (doc[i].text, doc[i+1].text) 
    if bigram in bigram_count:
        bigram_count[bigram] += 1
    else:
        bigram_count[bigram] = 1
print(bigram_count)
        
```

# Sampling a Sentence

```python
unigram_probabilities = {}
for unigram in unigram_count:
    unigram_probabilities[unigram] = unigram_count[unigram]/total_words_in_corpus
print(unigram_probabilities)

bigram_probabilities = {}
for bigram in bigram_count:
    count_prev_token = unigram_count[bigram[0]]
    bigram_probabilities[bigram] = bigram_count[bigram]/count_prev_token
print(bigram_probabilities)

uni_prob_keys = list(unigram_probabilities.keys())
uni_prob_values = list(unigram_probabilities.values())
```

```python
def generate_sentence():    
    global wordlist
    wordlist = []

    def sample_a_sentence(key_list, value_list):
        wordlist.append(random.choices(key_list, weights=value_list, k=1)[0])
        return wordlist[-1]

    sentence = sample_a_sentence(uni_prob_keys, uni_prob_values)

    while wordlist[-1] != ".":
        prev_word = wordlist[-1]

        bi_prob_keys = []
        bi_prob_values = []

        for key, value in bigram_probabilities.items():
            if key[0] == prev_word:
                bi_prob_keys.append(key[1])
                bi_prob_values.append(value)

        next_word = sample_a_sentence(bi_prob_keys, bi_prob_values)
        sentence = sentence + " " + next_word

    print(sentence)
```
